```python
import re
def str2array(s):
    s = s.replace("[", "").replace("]","").replace("\n", "")
    s =  re.sub(' +', ' ', s.strip())
    values = s.split(" ")
    arr = np.array([int(v) for v in values])
    return arr
```

```python
str2array(data[data.place=='Paris'].img.values[0])
```
